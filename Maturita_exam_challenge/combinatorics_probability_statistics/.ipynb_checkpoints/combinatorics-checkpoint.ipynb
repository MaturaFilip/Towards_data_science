- Rule of product / multiplication principle = if there are 'n' ways of doing something, and 'm' ways of doing another thing after that, then there are $n \times $m ways to perform both of thse actions (2 persons, 4 ice cream flavors - how many pairs I can create?)

- Addition principle / rule of sum = if there are $n$ choices for one action, and $m$ choices for another action and the two actions cannot be done at the same time, then there are $n+$m ways to choose one of these actions (3 north shops, 2 south shops -> need to decide which shop to visit  $3+2 = 5$ -> possible shops the person could end up shopping at today)

factorials:

- I have 4 people, in how many ways can they be put together to form a line?
- First position: 4 options
- Second position: 3 options
- Third: 2 options
- Fourth: 1 options

___4_, _3_, _2_, _1___

```python
num_people = 4
options = 1

for i in range(1, num_people + 1):
    options *= i

options
```

```python
# define factorial function
def factorial(num):
    counter = 1
    for i in range(1, num + 1):
        counter *= i
    return counter
```
